```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Cost function: J(w) = (w - 3)^2 + 2
def J(w):
    return (w - 3)**2 + 2

# Derivative of J
def dJ_dw(w):
    return 2 * (w - 3)
```

```python
# Gradient Descent
w = 10  # start far from min
alpha = 0.1
all_steps = [w]

for i in range(20):
    w = w - alpha * dJ_dw(w)
    all_steps.append(w)
```

```python
# Plotting
w_vals = np.linspace(-6, 12, 100)
plt.plot(w_vals, J(w_vals), label="Cost Function J(w)")
plt.scatter(all_steps, J(np.array(all_steps)), c='red', label="Steps")
plt.xlabel("w")
plt.ylabel("J(w)")
plt.legend()
plt.grid(True)
plt.show()
```
